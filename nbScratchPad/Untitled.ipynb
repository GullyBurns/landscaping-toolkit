```python
import os 
import pandas as pd
import re

path = '/Users/gburns/Documents/2023H1/Misdiagnsosis/2023-02-08'
files = os.listdir(path)
feature_frames = []
for f in sorted(files):
    if f[-4:]!='.tsv':
        continue
    m = re.search('^(.*)_(diag|proc|drug)_feature_summary.tsv', f)
    if m:
        c = m.group(1)
        t = m.group(2)
        df = pd.read_csv(path+'/'+f, sep='\t')
        for i, row in df.iterrows():
            f = row['feature']
            for a in range(18):
                #v = row[str(a+1)]
                v,e = row[str(a+1)].split(' +|- ')
                feature_frames.append((c,t,f,a+1,v,e))
#df
df = pd.DataFrame(feature_frames,columns=['cohort','type','feature','age','value','error'])
df
```

```python
df.pivot(index=['type','feature','cohort'],
               columns=['age'],
               values=['value']).to_csv(path+'/summary.tsv', sep='\t')
```

```python
pd.pivot_table(df,index=['type','feature','cohort'],
               columns=['age'],
               values=['value']).to_csv(path+'/summary2.tsv', sep='\t')
```
